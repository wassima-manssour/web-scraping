**e. Utiliser BeautifulSoup**

```python
from urllib import request

url = "https://www.bbc.com/news/world-latin-america-63167941"
htmlbbc = request.urlopen(url).read().decode('utf8')
from bs4 import BeautifulSoup

soup = BeautifulSoup(htmlbbc, 'html.parser')
title = soup.find('title')
title
```

## Scraping urlib et lxml

**a. Inspectez le code source de la page de BBC.<br>
b. Afficher les attribue de toute les balise** 

```python
from lxml import html

tree = html.fromstring(htmlbbc)
elems= tree.xpath('//a[@class="ssrcss-1j8v9o5-PromoLink e1f5wbog0"]')
for elem in elems:
    print(elem.attrib)
```

## Scraping avec Selenium

**a. Accéder à l’url https://www.tripadvisor.com/ <br>
b. Rechercher Marrakech <br>
c. Rechercher les hôtels de Marrakech<br>
d. Implémenter le scraper en selenium**

```python
pip install selenium
```

```python
from selenium import webdriver 
from selenium.webdriver.chrome.options import Options
import time

chrome_options = Options()
# chrome_options.headless = True # 
driver = webdriver.Chrome("chromedriver.exe",options=chrome_options)
```

```python
url = "https://www.tripadvisor.com/"
driver.get(url)
driver.get_screenshot_as_file("screenshot.png")
time.sleep(10)
XPath= "/html/body/div[1]/main/div[3]/div/div/div[2]/form/input[1]"
inputElement = driver.find_element("xpath",XPath)
inputElement.send_keys('Marrakech')

#inputElement.send_keys(Keys.ENTER) 
time .sleep(5)
text = "Tourism"
driver.find_element("xpath",'//a[contains(@href, "%s")]' % text).click()
time .sleep(5)
text = "Hotels"
driver.find_element("xpath",'//a[contains(@href, "%s")]' % text).click()
time .sleep(10)
```

**e. Récupérer les liens des hôtels de la première page**

```python
driver2 = webdriver.Chrome("chromedriver.exe",options=chrome_options)
url2 = "https://www.tripadvisor.com/Hotels-g293734-Marrakech_Marrakech_Safi-Hotels.html"
driver2.get(url2)
```

```python
from selenium.webdriver.common.by import By

links = driver2.find_elements(By.TAG_NAME, "a")
```

```python
links
```

```python
for link in links:
    print(link.get_attribute("href"))
```
